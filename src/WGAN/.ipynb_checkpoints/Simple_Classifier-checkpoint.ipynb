# Global variables

```python
image_size = 32
epochs = 30
batch_size = 64
show_step = 100
net = "VGG13"
tag = "Classifier"
tag = net + '_' + tag  
```

# Data Loader

```python
## Cracks and spalling dataset
din_path = "../data_in/ds1/train"
din_test_path = "../data_in/ds1/test"
```

```python
class load_Dataset(torch.utils.data.Dataset):
    def __init__(self, image_size, data_path=''):
        'Initialization'
        self.class_onehot={"nodmg":0, "cracks":1, "spalling":2}
        self.transform = transforms.Compose([transforms.Resize((image_size, image_size)), 
                                    transforms.ToTensor(),
                                    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))]
                                    )
        self.image_paths = glob.glob(data_path+'/*/*.jpg')
        
        
    def __len__(self):
        'Returns the total number of samples'
        return len(self.image_paths)
    
    def __getitem__(self, idx):
        'Generates one sample of data'
        image_path = self.image_paths[idx]
        image = Image.open(image_path)
        
        X = self.transform(image)
        y = torch.tensor(self.class_onehot[image_path.split('/')[-2]])
        return X, y
```

```python
## Load train and test data

train_dataset = load_Dataset(image_size, data_path=din_path)
test_dataset = load_Dataset(image_size, data_path=din_test_path)

train_dataloader = DataLoader(train_dataset, shuffle=True, batch_size=batch_size,
                              num_workers=1, drop_last=True)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size,
                                num_workers=1, drop_last=True)


print("Train set length: ", len(train_dataloader.dataset))
print("Test set length: ", len(test_dataloader.dataset))
```

# VGG-13 Model

```python
network = vgg13(num_classes=3)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(network.parameters())
```

```python
def test_metrics(test_loss, test_prec, test_rec):
    network.eval()
    
    prec_ = np.array([])
    rec_ = np.array([])
    loss_ = np.array([])
    for data, labels in tqdm(test_dataloader):
        outputs = network(data)
        loss__ = criterion(outputs, labels)

        pred = torch.argmax(outputs, dim=1)
        prec__, rec__ = precision_recall(pred, labels, average='macro', num_classes=3)
        
        prec_ = np.append(prec_, prec__.item())
        rec_ = np.append(rec_, rec__.item())
        loss_ = np.append(loss_, loss__.item())
        
    prec_ = np.mean(prec_)
    rec_ = np.mean(rec_)
    loss_ = np.mean(loss_)
        
    test_prec = np.append(test_prec, prec_)
    test_rec = np.append(test_rec, rec_)
    test_loss = np.append(test_loss, loss_)
        
    return test_loss, test_prec, test_rec

if not os.path.exists(f'../info/{tag}/'):
    os.mkdir(f'../info/{tag}/')
root_path=f'../info/{tag}/'

def save_checkpoint(epoch, test_loss, test_prec, test_rec):
    torch.save({
        'epoch': epoch,
        'test_losses': test_loss,
        'test_prec': test_prec,
        'test_rec': test_rec
        }, f"{root_path}tinfo.pkl")
    
    torch.save(network.state_dict(), f"{root_path}weights.pth")
  
    print("Saved checkpoint")
```

```python
def train(n_epochs):
    train_loss = np.array([])
    train_prec = np.array([])
    train_rec = np.array([])
    test_loss = np.array([])
    test_prec = np.array([])
    test_rec = np.array([])
    step = 0
    for epoch in range(n_epochs):
        print(f"------- Epoch {epoch+1} -------")
        for data, labels in tqdm(train_dataloader):
            # zero the parameter gradients
            network.train()
            optimizer.zero_grad()

            # forward + backward + optimize
            outputs = network(data)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            if (step%show_step == 0) or (step == 0):
                # Training metrics
                pred = torch.argmax(outputs, dim=1)
                prec_, rec_ = precision_recall(pred, labels, average='macro', num_classes=3)

                train_prec = np.append(train_prec, prec_)
                train_rec = np.append(train_rec, rec_) 
                train_loss = np.append(train_loss, loss.item())

                # Testing metrics
                test_loss, test_prec, test_rec = test_metrics(test_loss, test_prec, test_rec)
                try:
                    if (test_loss[-1]<test_loss[-2]): save_checkpoint(epoch, test_loss, test_prec, test_rec)
                except IndexError:
                    pass

                print(f"Train Loss: {train_loss[-1]:4f}  |  Test Loss: {test_loss[-1]:4f}")
                print(f"Train Prec: {train_prec[-1]:4f}  |  Test Prec: {test_prec[-1]:4f}")
                print(f"Train Rec:  {train_rec[-1]:4f}  |  Test Rec:  {test_rec[-1]:4f}")

                plt.plot(train_loss, label="Train Loss")
                plt.plot(test_loss, label="Test Loss")
                plt.legend()
                plt.show()

            step += 1
```

```python
train(epochs)
```
